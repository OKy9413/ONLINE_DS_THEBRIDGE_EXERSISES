```python
def get_companies(urls_paises = {
    "France": [
        "https://disfold.com/france/companies/?page=1",
        "https://disfold.com/france/companies/?page=2",
        "https://disfold.com/france/companies/?page=3",
        "https://disfold.com/france/companies/?page=4"
    ],
    "Germany": [
        "https://disfold.com/germany/companies/?page=1",
        "https://disfold.com/germany/companies/?page=2",
        "https://disfold.com/germany/companies/?page=3",
        "https://disfold.com/germany/companies/?page=4"
    ],
    "United-Kingdom": [
        "https://disfold.com/united-kingdom/companies/?page=1",
        "https://disfold.com/united-kingdom/companies/?page=2",
        "https://disfold.com/united-kingdom/companies/?page=3",
        "https://disfold.com/united-kingdom/companies/?page=4"
    ],
    "Spain": [
        "https://disfold.com/spain/companies/?page=1",
        "https://disfold.com/spain/companies/?page=2",
        "https://disfold.com/spain/companies/?page=3",
        "https://disfold.com/spain/companies/?page=4"
    ],
    "Italy": [
        "https://disfold.com/italy/companies/?page=1",
        "https://disfold.com/italy/companies/?page=2",
        "https://disfold.com/italy/companies/?page=3",
        "https://disfold.com/italy/companies/?page=4"
    ],
    "Ireland": [
        "https://disfold.com/ireland/companies/?page=1",
        "https://disfold.com/ireland/companies/?page=2"
    ]
} ):
    nombres_empresas = []
    paises_empresas = []

    for pais, urls in urls_paises.items():
        for url in urls:
            response = requests.get(url)
            soup = BeautifulSoup(response.text, 'html.parser')
            table = soup.find('table', {'class': 'striped'})
            rows = table.find_all('tr')

            for row in rows:
                cols = row.find_all('td')
                if len(cols) >= 2:
                    nombres_empresas.append(cols[1].get_text().strip())
                    paises_empresas.append(pais)
            
            time.sleep(1)  # Espera 1 segundo entre solicitudes

    return (nombres_empresas, paises_empresas)
```

```python
def comp_name(quantity, countries=None):
    # Si no se especifican países, se utilizan estos por defecto
    if countries is None:
        countries = ["Spain", "Italy", "Germany", "United-Kingdom", "Ireland", "France"]

    # Asumiendo que 'companies' es una tupla (lista_empresas, lista_paises)
    lista_empresas, lista_paises = companies

    # Filtrar las empresas por los países especificados
    empresas_filtradas = [(lista_empresas[i], lista_paises[i]) for i in range(len(lista_empresas)) if lista_paises[i] in countries]

    # Asegurarse de que la cantidad solicitada no sea mayor que la cantidad disponible
    quantity = min(quantity, len(empresas_filtradas))

    # Seleccionar aleatoriamente una cantidad determinada de empresas
    empresas_seleccionadas = random.sample(empresas_filtradas, quantity)

    # Crear y devolver un diccionario con los nombres de las empresas y sus países
    resultado = {nombre: pais for nombre, pais in empresas_seleccionadas}

    return resultado
```

```python
def remove_short_last_word(input_str):
    words = input_str.split()
    if words:
        last_word = words[-1]
        if len(last_word) <= 3:
            words.pop()
        return ' '.join(words)
    else:
        return input_str
```

```python
# Función para generar ID de proveedor
def generate_provider_id(company_name, country):
    company_name = remove_short_last_word(company_name)
    # Obtener las iniciales del nombre de la compañía
    words = company_name.split()
    if len(words) >= 3:
        initials = words[0][0] + words[1][0] + words[2][0]
    elif len(words) == 2:
        initials = words[0][:2] + words[1][0]
    elif len(words) == 1:
        initials = words[0][:3]
    else:
        # En caso de que no haya palabras en el nombre
        initials = "XXX"

    # Generar los 3 números aleatorios
    random_digits = ''.join(random.choice('0123456789') for _ in range(3))

    # Mapear el código del país
    country_codes = {
        'Spain': 'SP',
        'France': 'FR',
        'Germany': 'GE',
        'Ireland': 'IR',
        'United Kingdom': 'UK',
        'Italy': 'IT',
    }
    country_code = country_codes.get(country, 'XX')  # 'XX' para países desconocidos

    # Combinar las partes para formar el ID de proveedor
    provider_id = f'prov{initials}{random_digits}{country_code}'

    return provider_id
```

```python
def generar_id_pedidos(num_elementos):
    lista_id_pedidos = []
    for _ in range(num_elementos):
        pedido = 'pedi' + ''.join(random.choice('0123456789') for _ in range(6))
        lista_id_pedidos.append(pedido)
    return lista_id_pedidos
```

```python
def check_repetition(lst, target_element):
    threshold = len(lst) * 0.45
    count = 0
    
    for item in lst:
        if item == target_element:
            count += 1
            if count > threshold:
                return True
    
    return False
```

```python
def selec_prov_aleatorio_v2(num_elementos, df_prov = df_prov):
    # Número de proveedores diferentes
    if num_elementos <= len(df_prov['ID_proveedor']):
        n_prov = int(math.sqrt(num_elementos)) + 1
    else:
        n_prov = num_elementos
    
    # Obtener la lista de ID de proveedores
    id_proveedores = df_prov['ID_proveedor'].tolist()
    
    # Realizar la selección
    lista_seleccionada = random.sample(id_proveedores, n_prov)
    
    # Declaramos la lista prov_selec
    prov_selec = []

    # Calcular la cantidad de repeticiones y añadir los elementos a la lista
    prov_selec.extend(lista_seleccionada)
    while len(prov_selec) < num_elementos:
        sel_random = random.choice(lista_seleccionada)
        if check_repetition(prov_selec, sel_random):
            continue
        n_random = random.randint(1,int(math.sqrt(num_elementos - len(prov_selec))))
        if len(prov_selec) + n_random > num_elementos:          # Esta comrobación es un poco inutil ya que con la raiz cuadrada siempre se va cumplir. 
            continue
        for i in range(n_random):
            prov_selec.append(sel_random)
    
    return prov_selec
```

```python
def generar_fechas_aleatorias(num_elementos, num_meses = 24):
    # Obtener la fecha actual
    fecha_actual = datetime.now()

    # Calcular la fecha de inicio como 'num_meses' meses atrás
    fecha_inicio = fecha_actual - timedelta(days=num_meses * 30)

    # Generar fechas aleatorias dentro del rango especificado
    fechas_aleatorias = []
    for _ in range(num_elementos):
        # Calcular un número aleatorio de días entre 0 y 'num_meses * 30'
        dias_aleatorios = random.randint(0, num_meses * 30)

        # Calcular la fecha aleatoria dentro del rango
        fecha_aleatoria = fecha_inicio + timedelta(days=dias_aleatorios)

        # Formatear la fecha en "día mes año" y agregarla a la lista
        fecha_formateada = fecha_aleatoria.strftime("%d %m %Y")
        fechas_aleatorias.append(fecha_formateada)

    return fechas_aleatorias
```

```python
def generar_df_pedidos(num_filas, num_meses = 24):
    numero_filas = 10
    id_ped = generar_id_pedidos(num_filas)
    id_prov = selec_prov_aleatorio_v2(num_filas)
    date = generar_fechas_aleatorias(num_filas, num_meses)

    df_ped = pd.DataFrame({'ID_pedido': id_ped, 'ID_proveedor': id_prov, 'Fecha': date})

    df_ped['Fecha'] = pd.to_datetime(df_ped['Fecha'], format='%d %m %Y')
    df_ped.sort_values('Fecha', inplace=True, ascending= False)
    df_ped = df_ped.reset_index(drop = True)
    df_ped['Fecha'] = df_ped['Fecha'].dt.strftime("%d/%m/%Y")
    return df_ped
```

```python
def asignar_estado(fecha, estados, porcentajes = [0.4, 0.29, 0.15, 0.09, 0.05, 0.02]):
    fecha_actual = datetime.now()
    dias_diferencia = (fecha_actual - fecha).days

    if dias_diferencia <= 0:
        return estados[0]
    elif dias_diferencia > sum(porcentajes):
        return estados[-1]
    else:
        probabilidad_acumulada = 0
        for i, porcentaje in enumerate(porcentajes):
            probabilidad_acumulada += porcentaje
            if dias_diferencia <= probabilidad_acumulada:
                return estados[i]
```
